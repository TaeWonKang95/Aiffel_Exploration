# Exploration03 - 스티커 붙히기 - 원본

## 패키지 import

```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
import dlib
```

## 이미지 불러오기

```python
my_image_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/original3.png'
img_bgr = cv2.imread(my_image_path)    # OpenCV로 이미지를 불러옵니다
img_original = img_bgr.copy()      # 출력용 이미지를 따로 보관합니다
plt.imshow(img_bgr)
plt.show()
```

opencv에서는 색상을 BGR 순으로 사용하기 때문에 색상이 원본과 다르게 보임  
(matplotlib, dlib 등은 RGB 순으로 사용)

```python
# plt.imshow 이전에 RGB 이미지로 바꾸는 것을 잊지마세요. 
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

## 얼굴에서의 bounding box 추출

```python
# detector를 선언합니다
detector_hog = dlib.get_frontal_face_detector()
# dlib_rects = detector_hog(img_rgb, 1)
```

```python
# img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
dlib_rects = detector_hog(img_rgb, 1)   # (image, num of image pyramid)
```

```python
print(dlib_rects)   

for dlib_rect in dlib_rects:
    l = dlib_rect.left()
    t = dlib_rect.top()
    r = dlib_rect.right()
    b = dlib_rect.bottom()

    cv2.rectangle(img_original, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA)

img_original_rgb =  cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB)
plt.imshow(img_original_rgb)
plt.show()
```

## 랜드마크 불러오기

```python
model_path = os.getenv('HOME')+'/aiffel/camera_sticker/models/shape_predictor_68_face_landmarks.dat'
landmark_predictor = dlib.shape_predictor(model_path)
```

```python
list_landmarks = []

# 얼굴 영역 박스 마다 face landmark를 찾아냅니다
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)
    # face landmark 좌표를 저장해둡니다
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
    list_landmarks.append(list_points)

print(len(list_landmarks[0]))
```

```python
for landmark in list_landmarks:
    for point in landmark:
        cv2.circle(img_original, point, 2, (0, 255, 255), -1)

img_original_rgb = cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB)
plt.imshow(img_original_rgb)
plt.show()
```

## 스티커 붙히기

계산하기 쉽도록 bounding box의 좌표 표기  
- left : 200  
- top : 201  
- right : 468  
- bottom : 468  

### 1. 스티커 붙힐 좌표 파악

코 끝을 기준으로 문제를 풀어가려고 함
- 코 끝의 인덱스 번호 : 33

```python
for dlib_rect, landmark in zip(dlib_rects, list_landmarks):
    print (landmark[33])
    x = landmark[33][0]
    y = landmark[33][1]
    w = dlib_rect.width()
    h = dlib_rect.height()
    print ('(x,y) : (%d,%d)'%(x,y))
    print ('(w,h) : (%d,%d)'%(w,h))
```

### 2. 스티커 불러오고 사이즈 조정

```python
sticker_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/cat-whiskers.png'
img_sticker = cv2.imread(sticker_path) # 스티커 이미지를 불러옵니다
img_sticker = cv2.resize(img_sticker, (w,h))
print (img_sticker.shape)
```

사이즈는 bounding box의 w,h을 기준으로 조정함

### 3. 스티커 붙일 좌표 파악하기
- 스티커 이미지의 시작점은 top-left 좌표

- **좌표를 생각하기 어려워 노트에 직접 적어가며 파악함**
- 스티커의 크기는 bounding box의 크기와 같음
- 스티커의 중앙이 코끝 좌표에 와야함
- 스티커의 중앙을 bounding box의 좌측 상단으로 지정
- 좌표는 좌측 상단이 기준이므로 (-w//2, -h//2)
- 스티커의 중앙을 코끝의 좌표로 보냄
- 좌표는 좌측 상단이 기준이므로 (x-w//2, y-h//2)

```python
re_x = x - w // 2
re_y = y - h // 2
print ('(x,y) : (%d,%d)'%(re_x, re_y))
```

스티커의 좌측 상단 좌표 : (205, 223)

![좌표정리](camera_sticker/images/스티커정리.jpg)

### 4. 스티커가 이미지 범위를 넘어서는 영역 자름

```python
if re_x < 0: 
    img_sticker = img_sticker[:, -re_x:]
    re_x = 0
if re_y < 0:
    img_sticker = img_sticker[-re_y:, :]
    re_y = 0

print ('(x,y) : (%d,%d)'%(re_x, re_y))
```

### 5. 스티커의 영역을 설정

```python
img_sticker.shape
```

```python
sticker_area = img_original[re_y: re_y + img_sticker.shape[0], re_x: re_x+img_sticker.shape[1]]
img_original[re_y: re_y + img_sticker.shape[0], re_x: re_x+img_sticker.shape[1]] = \
    np.where(img_sticker==255, sticker_area, img_sticker).astype(np.uint8)
```

고양이 수염의 배경이 흰색이고 수염이 투명이므로 img_sticker==255로 지정

### 6. 스티커 위치 확인

```python
plt.imshow(cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB))
plt.show()
```

## 원본 이미지에 적용

```python
sticker_area = img_bgr[re_y: re_y + img_sticker.shape[0], re_x: re_x+img_sticker.shape[1]]
img_bgr[re_y: re_y + img_sticker.shape[0], re_x: re_x + img_sticker.shape[1]] = \
    np.where(img_sticker==255, sticker_area, img_sticker).astype(np.uint8)
plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.show()
```
